```python
# singular value decomposition(SVD) to rebuild the image that usess less singular values
from numpy import linalg
```

```python
# to get gray scale image

# Y = 0.2126R  + 0.7152G + 0.0722B
#         or
# img_gray = img_array @ [0.2126 , 0.7152, 0.0722]
```

```python
img_gray = img_array @ [0.01 , 0.0, 0.0]
```

```python
img_gray.shape
# no color in image
```

```python
plt.imshow(img_gray, cmap="gray")
plt.show()
```
